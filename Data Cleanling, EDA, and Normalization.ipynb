```python
columns_to_drop = [
    "Customer Email",
    "Customer Password",
    "Customer Street",
    "Product Image",
    "Order Zipcode",
    "Customer Zipcode"
]

# Drop only if the columns exist
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

print("Remaining columns:", len(df.columns))
```

```python
df = df.drop(columns=[
    "Order Customer Id",
    "Product Category Id",
    "Order Item Cardprod Id",
    "Sales per customer",
    "Latitude",
    "Longitude"
])
```

**Normalization**

```python
customers = df[[
    "Customer Id",
    "Customer Fname",
    "Customer Segment",
    "Customer City",
    "Customer State",
    "Customer Country"
]].drop_duplicates()
```

```python
departments = df[[
    "Department Id",
    "Department Name"
]].drop_duplicates()
```

```python
products = df[[
    "Product Card Id",
    "Product Name",
    "Product Price",
    "Category Id",
    "Department Id"
]].drop_duplicates()
```

```python
categories = df[[
    "Category Id",
    "Category Name"
]].drop_duplicates()
```

```python
shipping = df[[
    "Shipping Mode",
    "Delivery Status",
    "Days for shipment (scheduled)",
    "Days for shipping (real)",
    "Late_delivery_risk"
]].drop_duplicates().reset_index(drop=True)

shipping["Shipping ID"] = shipping.index + 1

shipping = shipping[[
    "Shipping ID",
    "Shipping Mode",
    "Delivery Status",
    "Days for shipment (scheduled)",
    "Days for shipping (real)",
    "Late_delivery_risk"
]]
```

```python
locations = df[[
    "Market",
    "Order Region",
    "Order Country",
    "Order State",
    "Order City"
]].drop_duplicates().reset_index(drop=True)

locations["Location ID"] = locations.index + 1

locations = locations[[
    "Location ID",
    "Market",
    "Order Region",
    "Order Country",
    "Order State",
    "Order City"
]]
```

```python
df = df.merge(
    shipping,
    on=[
        "Shipping Mode",
        "Delivery Status",
        "Days for shipment (scheduled)",
        "Days for shipping (real)",
        "Late_delivery_risk"
    ],
    how="left"
)
```

```python
df = df.merge(
    locations,
    on=[
        "Market",
        "Order Region",
        "Order Country",
        "Order State",
        "Order City"
    ],
    how="left"
)
```

```python
orders = df[[
    "Order Item Id",
    "Order Id",
    "Customer Id",
    "Product Card Id",
    "Shipping ID",
    "Location ID",
    "order date (DateOrders)",
    "shipping date (DateOrders)",
    "Order Status",
    "Sales",
    "Order Item Quantity",
    "Order Item Discount",
    "Order Item Discount Rate",
    "Order Item Total",
    "Benefit per order",
    "Order Profit Per Order",
    "Order Item Profit Ratio"
]]
```

```python
dates = pd.DataFrame({
    "Date": pd.concat([
        pd.to_datetime(df["order date (DateOrders)"], format='mixed').dt.date,
        pd.to_datetime(df["shipping date (DateOrders)"], format='mixed').dt.date
    ]).drop_duplicates().sort_values().reset_index(drop=True)
})

dates["Date"] = pd.to_datetime(dates["Date"])

dates["Date ID"] = dates.index + 1
dates["Year"] = dates["Date"].dt.year
dates["Quarter"] = dates["Date"].dt.quarter
dates["Month"] = dates["Date"].dt.month
dates["Month Name"] = dates["Date"].dt.month_name()
dates["Day"] = dates["Date"].dt.day
dates["Day Name"] = dates["Date"].dt.day_name()

dates = dates[[
    "Date ID",
    "Date",
    "Year",
    "Quarter",
    "Month",
    "Month Name",
    "Day",
    "Day Name"
]]

dates.to_csv(output_path + "Date.csv", index=False)
```
